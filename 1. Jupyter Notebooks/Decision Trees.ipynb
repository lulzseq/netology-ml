```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
X1 = np.random.rand(10,2)
y1 = np.random.randint(2, size=10)

X_train, X_test, y_train, y_test = train_test_split(X1, y1,test_size=0.25, random_state=1)

clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

dot = StringIO()
export_graphviz(clf, out_file=dot, filled=True, rounded=True, class_names=['0', '1'])
graph = pydotplus.graph_from_dot_data(dot.getvalue())
graph.write_png('tree.png')
Image(graph.create_png())
```
