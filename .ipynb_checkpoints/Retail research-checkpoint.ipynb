Поскольку это задача регрессии, выберем для оценки моделей основные метрики регрессии — MAE, RMSE и R$^2$score. 

MAE (средняя абсолютная ошибка) - представляет собой сумму различий между прогнозными и фактическими значениями. Это дает представление о том, насколько прогнозы были неправильными.

RMSE (cреднеквадратичное отклонение) - мера различий между значениями, предсказанными моделью или оценщиком и наблюдаемыми значениями.

````python
def fit_and_test(model, X_train, y_train, X_test, y_test):
    model = fit_model(model, X_train, y_train)
    return predict_and_test(model, X_test, y_test)

def fit_model(model, X_train, y_train):
    model.fit(X_train, y_train)
    return model

def predict_and_test(model, X_test, y_test):
    y_pred = model.predict(X_test)
    metrics_list = get_metrics(y_test, y_pred)
    print('MAE = {}, RMSE = {}'.format(metrics_list[0], metrics_list[1]))
    return model, metrics_list[0], metrics_list[1]

def get_metrics(y_tested, y_predicted):
    return get_mae(y_tested, y_predicted), get_rmse(y_tested, y_predicted)

def get_mae(y_tested, y_predicted):
    assert y_tested.values.mean()!=0
    return mean_absolute_error(y_true=y_tested.values, y_pred=y_predicted)/y_tested.values.mean()

def get_rmse(y_tested, y_predicted):
    return np.sqrt(mean_squared_error(y_tested.values, y_predicted))
````

Разделим данные.

````python
df_train_and_valid, df_test = train_test_split(df, random_state=42, test_size=0.3)
df_train, df_valid = train_test_split(df_train_and_valid, random_state=42, test_size=0.2)
print('df_train.shape: {}\ndf_valid.shape: {}\ndf_test.shape: {}'.format(df_train.shape, 
                                                                         df_valid.shape, df_test.shape))
````

Поскольку в датасете представлены данные разного масштаба, то нормализуем их с помощью StandardScaler. Основная идея заключается в том, чтобы стандартизировать (mean = 0 и standard deviation = 1) объекты перед применением методов машинного обучения.

````python
scaler = StandardScaler()

df_train[num_col] = scaler.fit_transform(df_train[num_col])
df_valid[num_col], df_test[num_col] = scaler.transform(df_valid[num_col]), scaler.transform(df_test[num_col])
````

Поделим на привычные X_train, y_test и т.п.

````python
df_train.describe().T
````

````python
drop_col = ['Item_Outlet_Sales', 'Outlet_Identifier', 'Item_Identifier']
````

````python
X_train, X_valid, X_test = df_train.drop(drop_col, axis=1), df_valid.drop(drop_col, axis=1),df_test.drop(drop_col, axis=1)
y_train, y_valid, y_test = df_train[['Item_Outlet_Sales']].squeeze(), df_valid[['Item_Outlet_Sales']].squeeze(), df_test[['Item_Outlet_Sales']].squeeze()

print('X_train.shape: {}\nX_valid.shape: {}\nX_test.shape: {}\nlen(y_train): {}\nlen(y_valid): {}\nlen(y_test): {}'.format(X_train.shape, X_valid.shape, X_test.shape, len(y_train), len(y_valid), len(y_test)))
````

Возьмем линейную модель, потренируем ее и выберем лучший вариант.

````python
model_list = []

model_linear = LinearRegression()
model_list.append(fit_and_test(model_linear, X_train, y_train, X_valid, y_valid))

model_lasso = Lasso()
model_list.append(fit_and_test(model_lasso, X_train, y_train, X_valid, y_valid))

model_ridge = Ridge()
model_list.append(fit_and_test(model_ridge, X_train, y_train, X_valid, y_valid))
````

Лучше себя показала LassoRegression. Запомним ее.

````python
model_lasso_final = model_list[1]
model_lasso_final
````

Попробуем бустинг.

````python
model_xgb = xgb.XGBRegressor(random_state=42)
````

````python
fit_and_test(model_xgb, X_train, y_train, X_valid, y_valid)
````

Результат лучше, чем у линейных моделей. Попробуем улучшить модель, используя разные параметры.

````python
params = {'n_estimators': list(np.linspace(1, 100, 10, dtype=int)),
          'max_depth': list(np.linspace(1, 50, 10, dtype=int)),
          'learning_rate':[0.0001, 0.001, 0.01, 0.1, 0.2, 0.3],
          'random_state': [42],
          'n_jobs': [-1]}
````

````python
grid_xgb = GridSearchCV(model_xgb, params, cv=10)
grid_xgb.fit(X_train, y_train)

print(grid_xgb.best_params_)
````

````python
model_xgb_final= xgb.XGBRegressor(n_estimators=34, max_depth=6, learning_rate=0.1, random_state=42, n_jobs=-1)
model_list.append(fit_and_test(model_xgb_final, X_train, y_train, X_valid, y_valid))
````

Стекинг.

````python
estimators = [
    ('Linear', LinearRegression()),
    ('Lasso', Lasso()),
    ('Ridge', Ridge()),
    ('XGBoost', model_xgb_final)]
````

````python
stacking_regressor = StackingRegressor(estimators=estimators, 
                                       final_estimator=model_xgb_final)
stacking_regressor.fit(X_train, y_train)
````

````python
y_pred_stacking_regressor = stacking_regressor.predict(X_test)
score_ensemble = cross_val_score(stacking_regressor, X_train, y_train, cv=10)
print('Score ensemble: {:.2f}'.format(score_ensemble.mean()))
````

````python
print('\tMAE\t\t\tRMSE')
print(get_metrics(y_test, y_pred_stacking_regressor))
````

````python
model_list.append(fit_and_test(stacking_regressor, X_train, y_train, X_valid, y_valid))
````

````python
for model in model_list:
    print(predict_and_test(model[0], X_test, y_test)[0], model[0])
    print()
    print()
````
